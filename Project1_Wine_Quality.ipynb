```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

*   accuracy_score → Overall correctness.
*   precision_score → Correct positive predictions out of all predicted positives.


*   recall_score → Correct positive predictions out of all actual positives.
*   f1_score → Harmonic mean of precision and recall.

*  confusion_matrix → Shows TP, TN, FP, FN.
*  classification_report → Displays all the metrics together.







```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
precision = precision_score(y_test, y_pred)
print("Precision:", precision)
```

```python
recall = recall_score(y_test, y_pred)
print("Recall:", recall)
```

```python
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print(classification_report(y_test, y_pred))
```

Conclusion

In this project, the Wine Quality dataset was analyzed and preprocessed successfully. A new binary target variable (quality_label) was created by classifying wines with quality greater than or equal to 7 as GOOD (1) and the rest as BAD (0).

Logistic Regression was trained both before and after applying StandardScaler, and its performance was evaluated using Accuracy, Precision, Recall, F1-score, and the Confusion Matrix. Scaling improved the model's performance slightly.

Three machine learning models—Logistic Regression, K-Nearest Neighbors (KNN), and Decision Tree—were compared. Logistic Regression achieved the highest accuracy (89.34%) and was selected as the best-performing model.

Hyperparameter tuning was performed using GridSearchCV, which further optimized the Logistic Regression model. Finally, feature importance analysis showed that features such as alcohol, volatile acidity, and sulphates had the strongest influence on predicting wine quality.

Overall, Logistic Regression proved to be the most suitable model for this binary wine quality classification task.
